```python
input_encoding = 'shift-jis'
output_encoding = 'utf_8_sig'
input_lang = 'ja'
output_lang = 'en'

translator = Translator()
data=pd.read_csv(dirname+'/sample.csv',encoding=input_encoding)
l_translate=[]
for i in data['target']:
    while True:
        try:
            dst = translator.translate(i, src=input_lang, dest=output_lang)
            l_translate.append(dst.text)
            break
        except Exception as e:
            translator = Translator()

data['translated']=l_translate
# data.to_csv(dirname+"/result.csv",encoding=output_encoding ,index=False)
```

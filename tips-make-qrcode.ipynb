```python
import qrcode
import os
```

## 🛠️Make QR code

```python
print("Type any words you want to convert to a QR Code: ", end="")

# Input  messages or a link by any ways
# qr_string = input()
qr_string = 'https://www.kaggle.com/yutodennou/make-qrcode/'

# Converting data into QR code
img = qrcode.make(qr_string)
file_name = "qr_code.png"
# Save the image
img.save(file_name)
current_dir = os.getcwd()
print("Save QR code image here: {0}/{1}".format(current_dir, file_name))
```
